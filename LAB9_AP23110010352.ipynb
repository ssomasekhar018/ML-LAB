```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load into DataFrame
data = pd.read_csv('iris.csv')
data.head()  # Display first few rows to verify
```

```python
# Scatter plot: Sepal.Length vs Sepal.Width
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Sepal.Length', y='Sepal.Width', hue='Species', data=data)
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
# Another: Petal.Length vs Petal.Width
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Petal.Length', y='Petal.Width', hue='Species', data=data, palette='Set2')
plt.title('Petal Length vs Petal Width by Species')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

```python
# Basic countplot of Species
plt.figure(figsize=(8, 6))
sns.countplot(x='Species', data=data)
plt.title('Count of Iris Samples by Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
# Boxplot: Sepal.Length by Species
plt.figure(figsize=(8, 6))
sns.boxplot(x='Species', y='Sepal.Length', data=data)
plt.title('Sepal Length Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```

```python
# Bar plot: Average Petal.Width by Species
plt.figure(figsize=(8, 6))
sns.barplot(x='Species', y='Petal.Width', data=data, errorbar=None)  # errorbar=None for no CI bars
plt.title('Average Petal Width by Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Width')
plt.show()
```

```python
plt.figure(figsize=(20, 12)).suptitle("Iris Dataset Dashboard", fontsize=20)

# Using a 2x3 subplot grid
plt.subplot(2, 3, 1)
sns.scatterplot(x='Sepal.Length', y='Sepal.Width', hue='Species', data=data)
plt.title('Sepal Length vs Width', fontsize=12)
plt.xlabel('Sepal Length', fontsize=12)
plt.ylabel('Sepal Width', fontsize=12)

plt.subplot(2, 3, 3)
sns.scatterplot(x='Petal.Length', y='Petal.Width', hue='Species', data=data, palette='Set2')
plt.title('Petal Length vs Width', fontsize=12)
plt.xlabel('Petal Length', fontsize=12)
plt.ylabel('Petal Width', fontsize=12)

# Countplot of Species (placed in the middle, like the original)
plt.subplot(1, 3, 2)  # Spans the middle
sns.countplot(x='Species', data=data)
plt.title('Count of Samples by Species', fontsize=12)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.subplot(2, 3, 4)
sns.scatterplot(x='Sepal.Length', y='Petal.Length', hue='Species', data=data, palette='Set1')
plt.title('Sepal vs Petal Length', fontsize=12)
plt.xlabel('Sepal Length', fontsize=12)
plt.ylabel('Petal Length', fontsize=12)

plt.subplot(2, 3, 6)
sns.scatterplot(x='Sepal.Width', y='Petal.Width', hue='Species', data=data, palette='Dark2')
plt.title('Sepal vs Petal Width', fontsize=12)
plt.xlabel('Sepal Width', fontsize=12)
plt.ylabel('Petal Width', fontsize=12)

plt.tight_layout()  # Adjust spacing
plt.show()
```

```python
# Line Plot Example (Adapted for Iris)
data_sorted = data.sort_values('Sepal.Length')  # Sort for a "trend-like" view
plt.figure(figsize=(8, 6))
plt.plot(data_sorted['Sepal.Length'], data_sorted['Petal.Length'], label='Petal Length', color='blue', linestyle='--', marker='o')
plt.title('Line Plot: Sepal Length vs Petal Length (Sorted)', fontsize=12)
plt.xlabel('Sepal Length (Sorted)', fontsize=12)
plt.ylabel('Petal Length', fontsize=12)
plt.legend()
plt.grid(True)  # Styling example
plt.show()
```

```python
# Simulated Dodged Countplot (Categorical-Categorical Adaptation)
data['Sepal Category'] = np.where(data['Sepal.Length'] > 6, 'Long Sepal', 'Short Sepal')  # Create a second category
plt.figure(figsize=(8, 6))
sns.countplot(x='Species', hue='Sepal Category', data=data, dodge=True)  # Dodged
plt.title('Dodged Countplot: Species by Sepal Category', fontsize=12)
plt.show()

# Stacked Version
plt.figure(figsize=(8, 6))
sns.countplot(x='Species', hue='Sepal Category', data=data, dodge=False)  # Stacked (no dodge)
plt.title('Stacked Countplot: Species by Sepal Category', fontsize=12)
plt.show()
```

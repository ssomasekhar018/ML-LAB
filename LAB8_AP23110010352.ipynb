```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Load iris.csv (assumed in same directory as notebook)
iris = pd.read_csv('iris.csv')

# Quick preview
print(iris.head())
print(iris.info())  # Check data types and missing values
print(iris['Species'].value_counts())  # Categorical frequency
print(iris['Sepal.Length'].describe())  # Numerical summary
```

```python
# Bar Chart for Species (Matplotlib)
species_counts = iris['Species'].value_counts()
plt.figure(figsize=(8, 5))
plt.bar(species_counts.index, species_counts.values, color='skyblue', edgecolor='black')
plt.title('Bar Chart: Frequency of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Countplot for Species (Seaborn)
plt.figure(figsize=(8, 5))
sns.countplot(x='Species', data=iris, palette='Set2', order=iris['Species'].value_counts().index)
plt.title('Countplot: Number of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Pie Chart for Species (Matplotlib)
plt.figure(figsize=(7, 7))
plt.pie(species_counts.values, labels=species_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Pie Chart: Proportion of Iris Species')
plt.show()
```

```python
# Histogram for Sepal.Length (Matplotlib)
plt.figure(figsize=(8, 5))
plt.hist(iris['Sepal.Length'], bins=20, color='green', edgecolor='black')
plt.title('Histogram: Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# KDE Plot for Sepal.Length (Seaborn)
plt.figure(figsize=(8, 5))
sns.kdeplot(iris['Sepal.Length'], fill=True, color='orange')
plt.title('KDE Plot: Density of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```

```python
# Boxplot for Sepal.Length (Seaborn)
plt.figure(figsize=(8, 5))
sns.boxplot(y=iris['Sepal.Length'], color='lightblue')
plt.title('Boxplot: Sepal Length with Outliers')
plt.ylabel('Sepal Length (cm)')
plt.tight_layout()
plt.show()
```
